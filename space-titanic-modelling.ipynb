```python
path='../input/spaceship-titanic'
```

```python
train_df=pd.read_csv(path+'/train.csv')
test_df=pd.read_csv(path+'/test.csv')
train_df.head()
```

```python
import null_func,eda_func,feature_engineering_func
```

```python
train_drop_nan,train_df=null_func.null_converter(train_df)
```

```python
test_df.isnull().sum()
```

```python
test_df['Spa'].fillna(test_df['Spa'].median(), inplace=True)
test_df['VRDeck'].fillna(test_df['VRDeck'].median(), inplace=True)
test_df['RoomService'].fillna(test_df['RoomService'].median(), inplace=True)
test_df['FoodCourt'].fillna(test_df['FoodCourt'].median(), inplace=True)
test_df['ShoppingMall'].fillna(test_df['ShoppingMall'].median(), inplace=True)
test = test_df.fillna(test_df.agg(lambda x: pd.Series.mode(x)[0], axis=0))
test['spending']=test['RoomService']+test['FoodCourt']+test['ShoppingMall']+test['Spa']+test['VRDeck']
```

```python
data=train_drop_nan
```

```python
eda_func.EDA_function(data)
```

```python
data=feature_engineering_func.feature_engineering_function(data)
test=feature_engineering_func.feature_engineering_function(test)
```

```python
data.shape
```

## Modelling

```python

from sklearn.model_selection import cross_val_score,train_test_split

from sklearn.neural_network import MLPClassifier
from IPython.display import display
from sklearn import metrics #accuracy measure
from sklearn.metrics import confusion_matrix #for confusion matrix
from sklearn.metrics import f1_score
from sklearn.metrics import PrecisionRecallDisplay

from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
train_X=data.drop('Transported',axis=1)
train_Y=data['Transported'].values.ravel()
# X=data.drop('Transported',axis=1)
# Y=data['Transported']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm      ## SVC,NuSVC,LinearSVC

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier


from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import StackingClassifier

from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

### Logistic Regression

```python
lrc=LogisticRegression()
parameters = {'penalty': ('l1', 'l2', 'elasticnet', 'none'),
              'C':[1, 10],
              'solver':('newton-cg','lbfgs','liblinear','sag','saga'),
              'max_iter':range(100,1000,100)}

rfc_grid = GridSearchCV(lrc, param_grid = parameters, cv = 5, scoring = 'accuracy', n_jobs= -1)
rfc_grid.fit(train_X, train_Y)
print('Best Parameters : ', rfc_grid.best_params_)
print('-'*50)
print('Best Accuracy : ', rfc_grid.best_score_)
```

```python
print('Best Parameters : ', rfc_grid.best_params_)
print('-'*50)
print('Best Accuracy : ', rfc_grid.best_score_)
```

```python
model=LogisticRegression(C=1,max_iter=300,penalty='l1',solver='liblinear')
model.fit(train_X,train_Y)
predict1=model.predict(test)
train_acc1=model.score(train_X,train_Y)
print("mean accuracy of train",train_acc1)
# print("accuracy of logistic regression",metrics.accuracy_score(predict1,test_Y))
scores = cross_val_score(model, train_X, train_Y, cv=5)
print("Cross_Val_Score Accuracy with a standard deviation %0.2f  is %0.5f" %(scores.std(),scores.mean()))
# f1_score_each_class=f1_score(test_Y, predict1, average=None)
# print("F1-Score of class 0 %0.6f and F1-Score of class 1  %0.6f" %(f1_score_each_class[0],f1_score_each_class[1]))
```

### SVM

```python
model= svm.SVC()
model.fit(train_X,train_Y)
predict2=model.predict(test)
train_acc2=model.score(train_X,train_Y)
print("mean accuracy of train",train_acc2)
# print("accuracy of SVC",metrics.accuracy_score(predict6,test_Y))
```

### KNN

```python
model= KNeighborsClassifier()
model.fit(train_X,train_Y)
predict3=model.predict(test)
train_acc3=model.score(train_X,train_Y)
print("mean accuracy of train",train_acc3)
# print("accuracy of KNeighborsClassifier",metrics.accuracy_score(predict9,test_Y))
model.get_params()
```

```python
model= DecisionTreeClassifier()
model.fit(train_X,train_Y)
predict4=model.predict(test)
train_acc4=model.score(train_X,train_Y)
print("mean accuracy of train",train_acc4)
pd.Series(model.feature_importances_,train_X.columns).sort_values(ascending=True).plot.barh()
plt.show()
```

```python
model= BaggingClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=10, random_state=0)
model.fit(train_X,train_Y)
predict5=model.predict(test)
print("Train accuracy of BaggingClassifier SVC",model.score(train_X,train_Y))
```

```python
model= ExtraTreesClassifier()
model.fit(train_X,train_Y)
predict6=model.predict(test)
print("Train accuracy of ExtraTreesClassifier",model.score(train_X,train_Y))
# print("Test accuracy of ExtraTreesClassifier",metrics.accuracy_score(predict20,test_Y))
pd.Series(model.feature_importances_,train_X.columns).sort_values(ascending=True).plot.barh()
plt.show()
```

```python
model= AdaBoostClassifier()
model.fit(train_X,train_Y)
predict7=model.predict(test)
print("Train accuracy of AdaBoostClassifier",model.score(train_X,train_Y))
# print("Test accuracy of AdaBoostClassifier",metrics.accuracy_score(predict21,test_Y))
pd.Series(model.feature_importances_,train_X.columns).sort_values(ascending=True).plot.barh()
plt.show()
## 0.7618 score on leaderboard
```

```python
model= GradientBoostingClassifier()
model.fit(train_X,train_Y)
predict8=model.predict(test)
print("Train accuracy of GradientBoostingClassifier",model.score(train_X,train_Y))
# print("Test accuracy of GradientBoostingClassifier",metrics.accuracy_score(predict22,test_Y))
pd.Series(model.feature_importances_,train_X.columns).sort_values(ascending=True).plot.barh()
plt.show()

## 0.78325 score on leaderboard
```

```python
model= MLPClassifier(solver='adam', alpha=0.001,learning_rate= 'adaptive',hidden_layer_sizes=(500, 100),random_state=1, max_iter=500)
model.fit(train_X,train_Y)
predict11=model.predict(test)
print("Train accuracy of MLPClassifier",model.score(train_X,train_Y))
# print("Test accuracy of MLPClassifier",metrics.accuracy_score(predict24,test_Y))
model.get_params()
```

```python
model= StackingClassifier(estimators=[('KNN',KNeighborsClassifier(n_neighbors=10)),
                                    ('RBF',svm.SVC(probability=True,kernel='rbf',C=0.5,gamma=0.1)),
                                    ('RFor',RandomForestClassifier(n_estimators=500,random_state=0)),
                                    ('LR',LogisticRegression(C=0.05)),
                                    ('DT',DecisionTreeClassifier(random_state=0)),
                                    ('GRA',GradientBoostingClassifier()),
                                    ('svm',svm.SVC())],final_estimator=AdaBoostClassifier())
model.fit(train_X,train_Y)
predict10=model.predict(test)
print("Train accuracy of StackingClassifier",model.score(train_X,train_Y))
```

```python
model= VotingClassifier(estimators=[('KNN',KNeighborsClassifier(n_neighbors=10)),
                                    ('RBF',svm.SVC(probability=True,kernel='rbf',C=0.5,gamma=0.1)),
                                    ('RFor',RandomForestClassifier(n_estimators=500,random_state=0)),
                                    ('LR',LogisticRegression(C=0.05)),
                                    ('DT',DecisionTreeClassifier(random_state=0)),
                                    ('Ada', AdaBoostClassifier()),
                                    ('GRA',GradientBoostingClassifier())],voting='soft')
model.fit(train_X,train_Y)
predict9=model.predict(test)
print("Train accuracy of VotingClassifier",model.score(train_X,train_Y))
# print("Test accuracy of VotingClassifier",metrics.accuracy_score(predict22,test_Y))
```
